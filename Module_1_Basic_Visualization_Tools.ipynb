# Gráfico de Barras

###Lembrando que dados podem ter dois tipos:
Dados qualitativos <- descrevem alguma qualidade, atributo  ou 
categoria. </br>
Ex: Nesse conjunto de dados, é o nome de um automóvel.
<br></br>
Dados quantitativos <- São medidos e descritos numericamente.
Ex: pressão, quilometragem e peso.
<br></br>
Obs: algumas variáveis aparentimente quantitativas podem ser 
interpretadas qualitativas dependendo do contexto.

```r
'exemplo de gráfic de barras com qplot()'
qplot(mtcars$cyl,    # dataset
      geom = "bar",    # definindo grafico em barra
      fill = I("blue"),    # preenchimento azul
      colour = I("red"),    # contorno vermelho
      xlab = "Cilindros",    # título eixo X
      ylab = "Número de Veículos"    # título eixo y
      )
```

# Histograma

```r
'histograma simples'
qplot(
  mtcars$hp,
  geom = "histogram"
)
```

```r
# histograma simples 
'selecionando um espaço de análise'
qplot(
  mtcars$hp,
  geom = "histogram",
  binwidth = 25
)
```

```r
# histograma simpmles 
'mudando a cor das bordas para entender melhor'
qplot(
  mtcars$hp,
  geom = "histogram",
  binwidth = 25,
  colour = I("black")
)
```

```r
# histograma simpmles 
'também podemos mudar o range que queremos visualizar'
qplot(
  mtcars$hp,
  geom = "histogram",
  binwidth = 25,
  colour = I("black"),
  xlim = c(50,250)
)
```

```r
# histograma simpmles 
'mudar os nomes dos eixos'
qplot(
  mtcars$hp,
  geom = "histogram",
  binwidth = 25,
  colour = I("black"),
  xlab = "Horsepower",    # nome do eixo x
  ylab = "Número de carros"    # nome do eixo y
)
```

```r
# histograma simples
'deixando o preenchimento totalmente transparante'
qplot(
  mtcars$hp,
  geom = "histogram",
  binwith = 25,
  colour = I("black"),
  xlab = "Hoursepower",
  ylab = "Número de carros",
  alpha = I(0)
      # coloquei '0' no alpha mas podia ter colocado
      # '0.5' ou algo assim para ficar meio transparante só'
)
```

```r
# histograma simples
'título no histograma'
qplot(
  mtcars$hp,
  geom = "histogram",
  binwith = 25,
  colour = I("black"),
  xlab = "Hoursepower",
  ylab = "Número de carros",
  alpha = I(0),
  main = "histograma"    # esse é o título
)
```

# Gráfico de Pizza

```r
'primeiro vamos fazer um gráfico em barras empilhadas'
bar <- ggplot(mtcars,
              aes(x=1, y=sort(mtcars$carb),fill = sort(mtcars$carb))
              ) + geom_bar(stat="identity")
print(bar)
```

```r
'para transformar o gráfico de barras em pizza usamos o método theta=y'

barp <- bar + coord_polar(theta='y')
print(barp)
```

```r
barp <- barp + theme(    # usando a função tema
  axis.line=element_blank(),    # classe axis edita o gráfico
  axis.text.x=element_blank(),
  axis.text.y=element_blank(),
  axis.ticks=element_blank(),
  axis.title.y=element_blank(),
  panel.background=element_blank()) +    # deixando o fundo transparente
  labs(y="Carburetors")    # mudando o título do gráfico


print(barp)
```
