## Seleção de Dataset
### Atenção*

tive que mudar um pouco o código de seleção de datasete, pois tinha algumas coisas desatualizadas pela versão

```r
#Selecionado nosso dataset
mtcars %>% 
  # atribuir nomes de linha a variáveis
  tibble::rownames_to_column(var = "group") %>% 
  # atribua cada variável -- car names -- às suas variáveis relacionadas
  mutate(across(-group, list(scales::rescale), .names = "rescaled_{col}")) %>% 
  # selecione quais dados plotar
  head(3) %>% select(1:10) -> mtcars_radar
```

```r
options(warn=-1)
ggradar(mtcars_radar)
```

```r
options(repr.plot.width=15, repr.plot.height=20, repr.plot.units="in")
ggradar(mtcars_radar)
```

# gráfico de Waffle 

```r
install.packages("ggplot2")
install.packages("waffle")    # para criar o gráfico de waffle
```

```r
library(ggplot2)
library(waffle)
```

```r
expenses <- c(`Saúde ($43,212)`=43212,
              `Educação ($113,412)`=113412,
              `Transporte ($20,231)`=20231,
              `Entretenimento ($28,145)`=28145)
```

```r
waffle_expenses <- waffle(expenses/1235, rows=5, size=0.3,
                          colors=c("#c7d4b5", "#a3aabd", "#a0d0de", "#97b5cf"),
                          title="Despesas domésticas imaginárias a cada ano",
                          xlab="1 square = $934")

waffle_expenses
```

```r
# podemos usar o options para deixar o gráfico maior
options(repr.plot.width=20, repr.plot.height=4, repr.plot.units="in")
waffle_expenses
```
